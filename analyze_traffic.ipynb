## 3 Use Seaborn with data analysis

```python
import seaborn as sns
import pandas as pd
```

### Load variable

```python
result_file = 'result'
if not os.path.exists('result'):
    os.mkdir('result')
    raise ValueError('no file in result')
    exit()
os.chdir('result/')
f = open(result_file, 'rb')
df = pickle.load(f)
f.close()
os.chdir('..')
```

### Generate data frame

```python
df = pd.DataFrame(columns = ['method','Cache Capacity' ,'Average Internal Traffic(requests)'])
for method in result.keys():
    for r in result[method].keys():
        #print({'method':method, 'Cache Capacity':r, 'Average Internal Traffic(requests)': mean(result[method][r][4:8])})
        df = df.append({'method':method,'Cache Capacity':r ,'Average Internal Traffic(requests)':mean(result[method][r][4:8])},ignore_index=True)
```

```python
df['Average External Traffic(requests)'] = 2000 - df['Average Internal Traffic(requests)']
```

```python
sns.lineplot(x = 'Cache Capacity', y = 'Average Internal Traffic(requests)', hue = 'method',style="method", markers = True, data = df)
plt.legend(title='Cache strategy', loc='upper right', labels=['Dynamic inter-AS POP', 'MPC', 'LCE', 'LCD'])
#plt.axis([0,40,0,6])
plt.grid()
#sns.lineplot(x = 'Cache Capacity', y = 'Average Internal Traffic(requests)', hue = 'method', data = df)
#plt.axis([0,40,0,2000])
## 保存
if not os.path.exists('result/'):
    os.mkdir('result/')
os.chdir('result')
plt.savefig('Average_Internal_Traffic.png')
os.chdir('../')
```
